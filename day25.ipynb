```python
from utils import read_lines
from collections import defaultdict

def parse_input(input_file):
    return [line.split(' ') for line in read_lines(input_file)]

def get_value(registers, a):
    try:
        return int(a)
    except ValueError:
        return registers[a]
    
def execute(program, registers):
    i = 0
    out_count = 0
    last_out = 1
    while i < len(program):
        instruction = program[i]
        match instruction[0]:
            case 'cpy':
                a, b = instruction[1:]
                
                registers[b] = get_value(registers, a)
                i += 1
            case 'inc':
                registers[instruction[1]] += 1
                i += 1
            case 'dec':
                registers[instruction[1]] -= 1
                i += 1
            case 'jnz':
                a, b = instruction[1:]
                if get_value(registers, a):
                    i += get_value(registers, b)
                else:
                    i += 1
            case 'out':
                o = get_value(registers, instruction[1])
                print(registers)
                if (last_out == 0 and o == 1) or (last_out == 1 and o == 0):
                    last_out = o
                    out_count += 1
                    if out_count > 100:
                        return True
                else:
                    return False
                i += 1
            case _:
                raise ValueError(f'illegal instruction {instruction}')
    return registers

def part1(input_file):
    program = parse_input(input_file)
    i = 1
    while i < 3000:
        print(i)
        registers = defaultdict(int)
        registers['a'] = i
        if execute(program, registers):
            return i
        else:
            i += 1

```
